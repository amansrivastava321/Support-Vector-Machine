```python
cat_col=['workclass','education','maritalstatus','occupation','relationship','race','sex','native']
```

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
for i in cat_col:
    tr_data[i]=label_encoder.fit_transform(tr_data[i])
    te_data[i]=label_encoder.fit_transform(te_data[i])
```

```python
train_X=tr_data.iloc[:,0:13]
train_Y=tr_data.iloc[:,13]
test_x=te_data.iloc[:,0:13]
test_y=te_data.iloc[:,13]
```

```python
model_rbf=SVC(kernel='rbf')
model_rbf.fit(train_X,train_Y)
train_pred_rbf=model_rbf.predict(train_X)
test_pred_rbf=model_rbf.predict(test_x)
train_rbf_acc=np.mean(train_pred_rbf==train_Y)
test_rbf_acc=np.mean(test_pred_rbf==test_y)
train_rbf_acc
test_rbf_acc
```
